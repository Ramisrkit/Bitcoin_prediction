```python

df['Volume']=pd.to_numeric(df['Volume'].str.replace(',',''),errors='coerce')
df['Market Cap']=pd.to_numeric(df['Market Cap'].str.replace(',',''),errors='coerce')
```

```python
import numpy as np
df = np.log1p(df)
```

```python

for i in df.columns:
  if pd.api.types.is_numeric_dtype(df[i]):
    q1=df[i].quantile(0.25)
    q3=df[i].quantile(0.65)
    iqr=q3-q1
    low=q1-1.5*iqr
    high=q3+1.5*iqr
    df[i]=df[i].clip(low,high)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(['Market Cap'],axis=1),df['Market Cap'],test_size=0.2,random_state=42)
```

```python
model=RandomForestRegressor(max_depth=2,bootstrap=True,min_samples_split=10,max_samples=0.30,n_estimators=100,max_features=0.8)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
r2_score(y_test,y_pred)
```
